```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import os
from pandas.tools.plotting import scatter_matrix

```

```python
df=pd.read_csv("index_results_92.csv")
df["inform_risk_r"]=9-df["inform_risk"]
```

```python
font = {'family' : 'serif', 
        'weight' : 'normal', 
        'size' : 22}
plt.rc('font', **font)
```

```python
os.makedirs("img",exist_ok=True)
```

```python
cols  = ["wb_risk",'wb_resilience','ndgain_risk','inform_risk_r']
coldict = dict(wb_risk="Risk\n(this paper)",wb_resilience="Resilience\n(this paper)",ndgain_risk="NDGain",inform_risk_r="InfoRM")
```

```python
a=scatter_matrix(df[cols].rename(columns=coldict), figsize=(15, 15), diagonal='name', s=80);
for i in range(0,4):
    plt.text(.5,.5,df[cols].rename(columns=coldict).columns[i],transform=a[i,i].transAxes,ha="center",va="center")
    a[i,0].get_yaxis().set_ticks([])
    a[i,0].get_yaxis().set_label_text("")
    a[-1,i].get_xaxis().set_ticks([])
    a[-1,i].get_xaxis().set_label_text("")
plt.savefig("comparison_of_indices.png",bbox_inches="tight")
```
